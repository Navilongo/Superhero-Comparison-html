```python
import pandas as pd
```

```python
heroes1 = "superhero_data.csv"
```

```python
heroes1_df = pd.read_csv(heroes1)
```

```python
heroes1_df.head()
```

```python
heroes1_df = heroes1_df.rename(columns={"creator":"Universe",
                                        "name":"Superhero",
                                        "real_name":"identity",
                                        "history_text":"backstory",
                                        "powers_text":"powers",
                                        "intelligence_score":"intelligence score",
                                        "strength_score":"strength score",
                                        "speed_score":"speed score",
                                        "durability_score":"durability score",
                                        "power_score":"power score",
                                        "combat_score":"combat score",
                                        "type_race":"race"})
heroes1_df
```

```python
heroes2_df = heroes1_df[["Universe","Superhero","identity","backstory","powers","gender","race","alignment","teams",
                         "intelligence score","strength score","speed score","durability score","power score","combat score"]]
heroes2_df.head()
```

```python
heroes2_df = heroes2_df.dropna()
heroes2_df
```

```python
heroes2_df.count()
```

```python
marvel_heroes_df = heroes2_df[heroes2_df["Universe"] == "Marvel Comics"]
```

```python
marvel_heroes_df.head()
```

```python
marvel_heroes_df = marvel_heroes_df.drop_duplicates(subset=['Superhero','identity']).copy()
marvel_heroes_df
```

```python
marvel_data = marvel_heroes_df.reset_index(drop=True)
marvel_data
```

```python
dc_heroes_df = heroes2_df[heroes2_df["Universe"] == "DC Comics"]
dc_heroes_df.head()
```

```python
dc_heroes_df = dc_heroes_df.drop_duplicates(subset=['Superhero','identity']).copy()
dc_heroes_df
```

```python
dc_data = dc_heroes_df.reset_index(drop=True)
dc_data
```

```python
marvel_data.to_csv('marvel.csv', sep=',')
dc_data.to_csv('dc.csv', sep=',')
```

```python
heroes2_df.to_csv('heroes.csv', sep=',')
```
